```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv('diabetes.csv')

# Explore the dataset
print(data.head())  # Display the first few rows of the dataset
data.describe()
```

```python
import numpy as np

# Define colors for each category
colors = np.where(data['Outcome'] == 0, 'blue', 'red')

# Plot the scatter plot with two different colors
plt.scatter(data['BloodPressure'], data['Insulin'], c=colors, cmap='viridis')
plt.xlabel('BloodPressure')
plt.ylabel('Insulin')
plt.title('Scatter Plot of Dataset')
plt.show()
```

```python
X = data.drop('Outcome', axis=1)  # Features
y = data['Outcome']  # Target variable

# Split the dataset into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.43, random_state=42)
```

```python
knn_classifier = KNeighborsClassifier(n_neighbors=5)

# Train the classifier
knn_classifier.fit(X_train, y_train)

# Predict on the test set
y_pred = knn_classifier.predict(X_test)

# Evaluate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
